```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('F1Score_55-66 - Grouping_Labels(score=0.25) (1).csv')
df.head()

print(5/2)
```

```python
models = np.array(np.unique(df.Model).tolist())
df_features = df[['Model', 'Labels','Train_Images', 'Train_F1']]
# df_features.sort_values('Train_F1')
y_set = np.array(df_features[['Train_F1']])
x_set = np.array(df_features[['Model']])

label = np.array(np.unique(df_features[['Labels']]))

plt.rcParams["figure.figsize"] = (30,10)
plt.rcParams['figure.facecolor']='white'
for x, y in zip(x_set, y_set):
    plt.scatter(x, y, s=100)
    
plt.ylim(-1, 1)
plt.title(f'Korelasi Model dan F1-Score Setiap Label - Training')
plt.xlabel('Model')
plt.ylabel('F1-Score')
plt.legend(label, loc='best')
plt.savefig(f'all_model_train.png')
plt.show()
```
